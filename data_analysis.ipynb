```python
import pandas as pd
from scipy.stats import pearsonr
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('final_df_BGM.csv')
```

```python
# коэффициент корреляции между двумя столбцами
corr, p_value = pearsonr(df['R1'], df['G1'])
print(f"Корреляция между R1 и G1: {corr:.3f}, p-value: {p_value:.3f}")
```

```python
# матрица корреляций для всех числовых столбцов
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(24, 16))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Матрица корреляций Пирсона")
plt.show()
```

```python
corr, p_value = pearsonr(df['year'], df['madness'])
print(f"Корреляция между годом и madness: {corr:.3f}, p-value: {p_value:.3f}")
```
