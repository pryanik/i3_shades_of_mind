```python
import pandas as pd
from dotenv import load_dotenv
import os
```

```python
load_dotenv()

csv_lines_circles = os.getenv('csv_lines_circles')
csv_KMeans = os.getenv('csv_KMeans')
csv_BGM = os.getenv('csv_BGM')
csv_GM = os.getenv('csv_GM')
```

```python
csv_lines_circles = pd.read_csv(csv_lines_circles)
csv_lines_circles['year'] = csv_lines_circles['year'].astype(int)
csv_lines_circles.head()
```

```python
csv_KMeans = pd.read_csv(csv_KMeans)
csv_KMeans = pd.merge(csv_KMeans, csv_lines_circles, on='year', how='inner')
csv_KMeans.head()
```

```python
csv_KMeans['madness'] = [i / len(csv_KMeans) for i in range(1, len(csv_KMeans) + 1)]
```

```python
csv_KMeans.head()
```

```python
csv_KMeans.to_csv('final_df_KMeans.csv', index=False)
```

```python
csv_BGM = pd.read_csv(csv_BGM)
csv_BGM = pd.merge(csv_BGM, csv_lines_circles, on='year', how='inner')
csv_BGM.head()
```

```python
csv_BGM['madness'] = [i / len(csv_BGM) for i in range(1, len(csv_BGM) + 1)]
```

```python
csv_BGM.to_csv('final_df_BGM.csv', index=False)
```

```python
csv_GM = pd.read_csv(csv_GM)
csv_GM = pd.merge(csv_GM, csv_lines_circles, on='year', how='inner')
csv_GM.head()
```

```python
csv_GM['madness'] = [i / len(csv_GM) for i in range(1, len(csv_GM) + 1)]
```

```python
csv_GM.to_csv('final_df_GM.csv', index=False)
```
